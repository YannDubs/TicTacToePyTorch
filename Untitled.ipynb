```python
import pandas as pd
import os

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/tic-tac-toe/tic-tac-toe.data'
filename = "data.csv"
file = filename if os.path.isfile(filename) else url
df = pd.read_csv(file, header=None)
df.to_csv(filename, header=None, index=False)
```

```python
o_won = df.loc[df[9] == "negative", 0:8]
o_won["label"] = 1
x_won = o_won.replace({"x":"o", "o":"x"})
x_won["label"] = 1
ox_won_draw = pd.concat([o_won,x_won]).reset_index(drop=True)
df = df.iloc[:,0:9]
```
